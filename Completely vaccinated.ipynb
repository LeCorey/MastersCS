```python
import pandas as pd

# Path to the Excel file
file_path = '/Users/lecoreyjohnson/Downloads/COVID-19 County Vaccine Data-2 (2).xlsx'

# Load the Excel file
vaccine_data = pd.read_excel(file_path)

# Filter out the row for the entire state of Alabama to focus only on counties
county_data = vaccine_data[vaccine_data['County'] != 'Alabama']

# Calculate the vaccination rate for each county
county_data['Complete Vaccination Rate (%)'] = (county_data['People Completely Vaccinated'] / county_data['Population > 16']) * 100

# Sort the counties by the complete vaccination rate in descending order
sorted_counties = county_data.sort_values(by='Complete Vaccination Rate (%)', ascending=False)

# Set option to display all rows
pd.set_option('display.max_rows', None)

# Display the sorted data
sorted_counties[['County', 'Complete Vaccination Rate (%)']]

```
