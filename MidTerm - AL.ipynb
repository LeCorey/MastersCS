```python
import pandas as pd


file_path = '/Users/lecoreyjohnson/MidTerm - AL/COVID-19 in Alabama-2 (3).xlsx'

# Read the Excel file
data = pd.read_excel(file_path)

# Remove rows with missing data
data_cleaned = data.dropna()

# Calculate death rates: (Deaths / Cases) * 100
data_cleaned['Death Rate'] = (data_cleaned['Deaths'] / data_cleaned['Cases']) * 100

# Sort the counties by death rate in descending order
sorted_data = data_cleaned.sort_values(by='Death Rate', ascending=False)

# Set pandas to display all rows
pd.set_option('display.max_rows', None)

# Display the sorted data
print(sorted_data[['Counties', 'Death Rate']])
```
