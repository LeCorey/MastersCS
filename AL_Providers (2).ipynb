```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the provided file path
file_path = '/Users/lecoreyjohnson/Downloads/COVID-19 in Alabama Vaccine Providers-2.csv'

try:
    df = pd.read_csv(file_path)
except UnicodeDecodeError:
    # If the default encoding doesn't work, try 'ISO-8859-1'
    df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Group the data by county and count the number of providers
county_vaccine_providers = df.groupby('County').size().reset_index(name='Number of Providers')

# Sort the data by county name for better readability
county_vaccine_providers_sorted = county_vaccine_providers.sort_values(by='County')

# Create a bar chart
plt.figure(figsize=(15, 8))  # Adjust the size as needed
plt.bar(county_vaccine_providers_sorted['County'], county_vaccine_providers_sorted['Number of Providers'], color='skyblue')
plt.xlabel('County')
plt.ylabel('Number of Vaccine Providers')
plt.title('Number of Vaccine Providers by County in Alabama')
plt.xticks(rotation=90)  # Rotate labels for better readability
plt.tight_layout()  # Adjust layout to fit all labels
plt.show()
```
